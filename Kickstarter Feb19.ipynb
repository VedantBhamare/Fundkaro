```python
#converting categorical values to numeric values
from sklearn.preprocessing import LabelEncoder
Categorical_var=['name','currency','main_category','sub_category','status']
le=LabelEncoder()
for i in Categorical_var:
    df[i] = le.fit_transform(df[i])
df.head()
```

```python
df.info()
```

```python
X=df.drop(['id','launched_at','launched_at','deadline','city','state','country','status','start_Q','end_Q'],axis=1)
Y=df['status']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=110)
```

```python
#Remeber to scaler the data
from sklearn.linear_model import LogisticRegression
my_model=LogisticRegression()
result=my_model.fit(X_train,Y_train)
```

```python
#Scaling the data
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
my_model=LogisticRegression()
result=my_model.fit(X_train,Y_train)
```

```python
prediction=result.predict(X_test)
prediction
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,prediction)
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

confusion_mat=confusion_matrix(Y_test,prediction)
```

```python
confusion_df=pd.DataFrame(confusion_mat,index=['Actual Neg','Actual Pos'], columns=['Predicted Neg','Predicted Pos'])
confusion_df
```

```python
color_conf_matrix=sns.heatmap(confusion_df,cmap='coolwarm',annot=True)
confusion_df
```
